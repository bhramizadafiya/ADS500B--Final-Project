2. Clean, Wrangle, and Handle Missing Data
Missing Values
Identify missing values using:

```python


print(data.isnull().sum())
```

```python
data['job'].replace('unknown', 'Not Specified', inplace=True)
data['education'].replace('unknown', 'Not Specified', inplace=True)
```

```python
data['contact'].fillna('unknown', inplace=True)
```

Numeric Columns: Use mean/median imputation or remove rows with missing data (depending on impact).



```python
data['balance'].fillna(data['balance'].median(), inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['balance', 'duration', 'campaign']] = scaler.fit_transform(data[['balance', 'duration', 'campaign']])
```

```python
data['quarter'] = data['month'].apply(lambda x: 'Q1' if x in ['jan', 'feb', 'mar'] else 'Q2' if x in ['apr', 'may', 'jun'] else 'Q3' if x in ['jul', 'aug', 'sep'] else 'Q4')
```

```python
data['age_group'] = pd.cut(data['age'], bins=[0, 30, 50, 100], labels=['Young', 'Middle-aged', 'Senior'])
```
